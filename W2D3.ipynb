## Inheritance

### #if child class have its own attributes it cannt use attributes of parent class but we can do it with super function

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera

class SmartPhone(Phone)      :
    def __init__(self,price,brand,camera,os,ram):
        super().__init__(price,brand,camera)
        self.os=os
        self.ram=ram
        print("inside smartphone constructor")

s=SmartPhone(20000,"samsung",12,"android",2)

print(s.os)
print(s.brand)
        
```

```python
class sensor:

    def __init__(self,location):
        self.location=location
        self.reading=[]
    def record(self,value):
        self.reading.append(value)
    
    def average(self):
        average=sum(self.reading)/len(self.reading)
        return average
    
    def latest(self):
        return self.reading[::-1]
    
    def str(self):
        return self.location
    
class thermometer(sensor):
    def __init__(self,unit,location):
        super().__init__(location)
        self.unit=unit


    def converter(self):
        if self.unit==  "centigrade":
            print (f"temp in fahrenheit = {self.average()*(9/5)+32} F ")
        elif self.unit=="fahrenheit":
            print(f"temparature in centigrade {self.average()-32*5/9}")   
```
