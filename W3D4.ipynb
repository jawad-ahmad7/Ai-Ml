```python
class Counter:
    def __init__(self,value=0):
        self.value=value

    def increment(self):
        self.value+=1
        return self.value
    
    def decrement(self):
        if self.value>0:
           if self.value==1:
               return f"warning! value deremented {self.value-1}"
           return f"value decremented by 1 new value {self.value-1}"
        else :
            print("value must be greater than 0")

    def reset(self):
        return self.value==0        

```

```python
c=Counter(10)
```

```python
c.increment()
```

```python
c.decrement()
```

```python
c.reset()
```

### TASK 2

```python
class Bird:

    def flight(self):
        print("i can fly")  

class sparrow(Bird):
    def chirp(self):
        print("sparrow is chirping")

s=sparrow()       
    
```

```python
s.flight()
```

```python
s.chirp()
```

```python

# Method overriding of flight
class penguin(Bird):
    def flight(self):
        print("sorry I cannot fly")
```

```python
p=penguin()
```

```python
p.flight()
```
