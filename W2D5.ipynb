```python
def calculator():
    num1=int(input("enter the first number "))
    num2=int(input("enter the seccond number"))
    op=input("enter operator")
    result =0
    if op=='+':
        result=num1+num2
    elif op=='-':
        result=num1-num2
    elif op=='*':
        result=num1*num2  
    elif op=='/':
        if num2 ==0:
            print("please enter a valid operator ")
        else:    
         result=num1/num2     
    else:
        print("enter  valid operator supported")
    print (f"the result is : {result}"  )    

```

```python
calculator()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Input (hours studied)
X = np.array([[1], [2], [3], [4], [5]])

# Output (exam scores)
y = np.array([45, 50, 58, 65, 72])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict score for 6 hours of study
prediction = model.predict([[6]])

print(prediction)
```
