#This section is all about file handling and file operations 

```python
#reading a whole file 

with open ('example.txt','r') as file:
    read=file.read()
    print(read)
```

```python
#reading a file line by line 
with open('example.txt','r') as file:
    for lines in file:
        print(lines)
```

```python
#removing line breaks in above text 
with open('example.txt','r') as file:
    for lines in file:
        print(lines.strip())
```

```python
# writing a file (overwritting)
with open('example.txt','w') as file:
    file.write('Hello everyone \n')
    file.write('i am abdullah')
    
with open('example.txt','r') as file:
    contents=file.read()
    print(contents)
```

```python
#writing a file without overwritting it 
with open('example.txt','a') as file:
    file.write('\n This is the third line \n')
    file.write('\n This is the fourth line')
with open('example.txt','r') as file:
    contents=file.read()
    print(contents)
```

```python
#writing multiple lines or list of lines 
list_of_line=['this is fifth line','this is sixth line','this is seventh line']
with open('example.txt','a') as file:
    file.writelines(list_of_line)
with open('example.txt','r') as file:
    contents=file.read()
    print(contents)
```

```python
#working with binary files 
data=b'\x00\x01\x02\x03\x04'
with open ('example.bin','wb') as file:
    file.write(data)
```

```python
#reading a binary file
with open('example.bin','rb') as file:
    contents=file.read()
    print(contents)
```

```python
#reading contents from source file and writing it destination file 
with open('example.txt','r')as file:
    contents=file.read()
    
with open('destination.txt','w')as file:
    file.write(contents)
    
with open('destination.txt','r')as file:
    content1=file.read()
    print(content1)
```

```python
#writing and then reading a file 
#first we will write in file 
list_of_line=['this is first line \n','this is 2nd line \n','this is 3rd line \n']
with open('new_file.txt','w+')as file:
    file.writelines(list_of_line)
    file.seek(0) # moves the cursor to start
    contents=file.read()
    print(contents)
```

```python
#creating a new directory 
import os
new_directory='new_folder'
a=os.mkdir(new_directory)
print(a)
```

```python
#listing files and directory in the current directory you are working 
items=os.listdir('.')
print(items)
```

```python
#joining paths 
dir_name='new_folder'
file_name='new_file.txt'
a=os.path.join(dir_name,file_name)
print(a)
```

```python
#path exsistence check 
path='destination.txt'
if os.path.exists(path):
    print(f'{path} exists')
else:
    print(f'{path} donot exists')
```

```python
#checking if path is directory or file 
path='example.bin'
if os.path.isfile(path):
    print(f'{path} is a file')
elif os.path.isdir(path):
    print(f"{path} is a directory")
else:
    print(f"{path} is neither a file nor a directory")
```

```python
#getting absolte path 
relative_path='example.txt'
absolute_path=os.path.abspath(relative_path)
print(absolute_path)
```
