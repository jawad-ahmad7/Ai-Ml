This section is all about seaborn.Seaborn is based on matplotlib and it is a visualization library .
It provides high level interface for drawing atractive and informative stastical graphics.
It provides complex visualization with few lines of code.


```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#creating a scatterplot 
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title("scatter plot of total_bill vs tip")
```

```python
tips
```

```python
#creatng a line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.xlabel('size')
plt.ylabel('total bill')
plt.title('size vs total bill')
```

```python
#categorical plots 
#1 Bar plot 
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("bar plot of day vs total bill")
```

```python
tips
```

```python
#box plot 
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#in case we just want that smoothing line we have KDE plot 
sns.kdeplot(x='total_bill',data=tips,fill=True)
```

```python
#pairplot
sns.pairplot(tips)
```

```python
tips
```

```python
#heatmap for corelation
correlation_matrix = tips[['total_bill', 'size', 'tip']].corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.show()
```
