This section is all about matplotlib.Matplolib is a powerful visualization library.

```python
import matplotlib.pyplot as plt
```

```python
#creating a line plot 
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.plot(x,y)
```

```python
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('LINE PLOT')
```

```python
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.plot(x,y,color='red',marker='o',linestyle='-',linewidth=2,markersize=5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('LINE PLOT')
```

```python
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.plot(x,y,color='red',marker='o',linestyle='-',linewidth=2,markersize=5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('LINE PLOT')
plt.grid(True)
```

```python
#creating multiplots 
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]
plt.figure(figsize=(9,5)) #9 inch width and 5 inch height

plt.subplot(1,2,1)
plt.plot(x,y1,color='yellow')
plt.title('first graph')

plt.subplot(1,2,2)
plt.plot(x,y2,color='red')
plt.title('2nd graph')

plt.show()
```

```python
#creating multiplots 
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]
plt.figure(figsize=(9,5)) #9 inch width and 5 inch height

plt.subplot(2,2,1)
plt.plot(x,y1,color='yellow')
plt.title('first graph')

plt.subplot(2,2,2)
plt.plot(x,y2,color='red')
plt.title('2nd graph')

plt.subplot(2,2,3)
plt.plot(x,y1,color='yellow')
plt.title('Third graph')

plt.subplot(2,2,4)
plt.plot(x,y2,color='red')
plt.title('fourth graph')

plt.show()
```

```python
#Creating a bar plot
categories=['A','B','C','D']
values=[2,3,4,5]
plt.bar(categories,values,color='purple')
```

```python
#creating histograms 
#Histograms are powerful plots that shows data distribution
data=[1,2,2,3,3,4,4,5,5,5]
plt.hist(data,bins=5,color='purple',edgecolor='black')
```

```python
#creating a scatter plot
x=[1,2,3,4,5]
y=[2,3,5,6,9]
plt.scatter(x,y,color='blue')
```

```python
#pie chart
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'blue']
explode = (0.2, 0, 0, 0) # The first slice ('A') moves outward from the center by 20% of the radius.

# FIX: Removed quotes and used 'colors=' instead of 'color='
plt.pie(sizes, labels=labels, colors=colors, explode=explode,autopct="%1.1f%%")

plt.show()
```
