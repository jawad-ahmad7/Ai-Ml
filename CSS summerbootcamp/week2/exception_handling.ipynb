Exception handling allows you to handle errors gracefully without stopping execution of porogram.

What are exception?

Exceptions are events that disrupt normal flows of porograms.They occur when error is encountared during porogram execution

Examples include 

Zero Division Error 

File Not Found Error

Value Error
 
Type Error

```python
try:
    a=b
except:
    print('This is an error')
```

```python
try:
    a=b
except NameError as ex:
    print(ex)
    
```

```python
try:
    result=1/0
except ZeroDivisionError as ex:
    print(ex)
    print('plz enter number greater than zero')
```

```python
try:
    a=b
    result=1/0
except ZeroDivisionError as ex:
    print(ex)
except Exception as ex:
    print(ex)
```

```python
try:
    num=int(input('please enter a number'))
    result=10/num
except ZeroDivisionError:
    print(f'value should be greater than 0')
except ValueError:
     print(f'value should be a number')
except Exception as ex:
    print(ex)
        
```

```python
#try,except,else
try:
    num=int(input('please enter a number'))
    result=10/num
except ZeroDivisionError:
    print(f'value should be greater than 0')
except ValueError:
     print(f'value should be a number')
except Exception as ex:
    print(ex)
else:
    print(f'result is = {result}')
```

```python
#try,except,else,finally
try:
    num=int(input('please enter a number'))
    result=10/num
except ZeroDivisionError:
    print(f'value should be greater than 0')
except ValueError:
     print(f'value should be a number')
except Exception as ex:
    print(ex)
else:
    print(f'result is = {result}')
finally:
    print('You need to handle the exceptions')
```

```python
#file handling and exception handling 
try:
    file=open('example.txt')
    contents=file.read()
except FileNotFoundError as ex:
    print(ex)
    print('file not found')
finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print('fil is closed now')
```
