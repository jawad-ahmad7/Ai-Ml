```python
word = "Python"
char_list = list(word)
print("Characters:", char_list)
```

```python
# Convert range to list
range_list = list(range(1, 6))
print("Range list:", range_list)
```

```python
# Convert tuple to list
tuple_data = (10, 20, 30)
list_data = list(tuple_data)
print("From tuple:", list_data)

```

```python
# Change single element
fruits = ["apple", "banana", "cherry"]
print("Original:", fruits)

fruits[1] = "blueberry"
print("After change:", fruits)
```

```python
# Change multiple elements using slicing
numbers = [1, 2, 3, 4, 5]
print("Original numbers:", numbers)

numbers[1:3] = [20, 30]
print("After slice change:", numbers)

```

```python
fruits = ["apple", "banana"]
print("Original:", fruits)

fruits.append("cherry")
print("After append:", fruits)
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

print("List 1:", list1)
print("List 2:", list2)

list1.extend(list2)
print("After extend:", list1)

print()
```
