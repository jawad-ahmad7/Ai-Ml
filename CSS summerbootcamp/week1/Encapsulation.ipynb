Encapsulation

Concept of wrapping data(variables) and methods together as a single unit 

Basically it restricts direct aceess to some of object components to prevent data misuse

```python
# involves concepts of getter and setters
# involves concept of public,private and protected access modifiers 
```

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age       # pubic instance variables 
        
def get_name(person):       #easily accessible see 
    return person.name

person=Person("Abdullah",22)
print(person.name)          ## see 
```

```python
# now using private 
# see private access modifier cannot be accessed outside the class 
# Derived class cannot even access it 
# Well we can accesss it inside that paticular class in which it is defined using getter and setters 
```

```python
class Person:
    def __init__(self,name,age,gender):
        self.__name=name
        self.__age=age  # private instance variables 
        self.gender=gender
        
def get_name(person):       #not  accessible see 
    return person.__name

person=Person("Abdullah",22,"MALE")
print(person.name)     #see 
```

```python
#moving toward how to access private acess modifiers in same class using getter and setters 
class Person:
    def __init__(self,name,age,gender):
        self.__name=name
        self.__age=age  # private instance variables 
        self.gender=gender
    
    def get_name(self):
        return self.__name
       
    def set_name(self,name):     
        self.__name=name
        
    

person=Person("Abdullah",22,"MALE")
print(person.get_name())
person.set_name("Ali")
print(person.get_name())
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name        
        self.__age = age         
        self.gender = gender

class Employee(Person):
    def display_details(self):
        print(f"Employee Name: {self.__name}") 


emp = Employee("Abdullah", 22, "MALE")


emp.display_details()
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name       
        self.__age = age          
        self.gender = gender

    def get_name(self):
        return self.__name       

class Employee(Person):
    def display_details(self):
        parent_name = super().get_name()
        print(f"Employee Name via super(): {parent_name}")


emp = Employee("Abdullah", 22, "MALE")


emp.display_details()  
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

   
    def __validate_age(self):
        if self.age < 0:
            return "Invalid Age"
        return "Valid Age"

    
    def display_profile(self):
        status = self.__validate_age()
        print(f"Name: {self.name}, Status: {status}")


person = Person("Abdullah", 22)


person.display_profile()  
```

```python
#lets move towards protected now 
#protected cannot be accessed outside the class but can be accessed through derived class 
```

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age
        
class Employe(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age)
        self.gender=gender
            
employe=Employe('abdullah',22,'MALE')
print(employe._name)            
```
