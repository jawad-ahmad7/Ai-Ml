Inheritance is a fundamental concept in OOP
It allows a class to inherit the attributes and methods of another class or parent class 
Types
1. Single Inheritance 
2. Multiple Inheritance 

```python
class Car:
    def __init__(self,windows,doors,engine_type):
        self.windows=windows
        self.doors=doors
        self.engine_type=engine_type
        
    def drive(self):
        print(f"The person will be driving {self.engine_type} car ")
        
class Tesla(Car):
    def __init__(self,windows,doors,engine_type,is_selfdrive):
        super().__init__(windows,doors,engine_type)
        self.is_selfdrive=is_selfdrive
    def selfdriving(self):
        print(f"Tesla supports selfdriving {self.is_selfdrive}")
        

            
```

```python
#creating objects
car1=Car(4,4,"Petrol")
print(car1.windows)
print(car1.doors)
car1.drive()
```

```python
#creating object of subclass/child class 
tesla1=Tesla(2,2,"Electric",True)
print(tesla1.windows)
print(tesla1.doors)
print(tesla1.is_selfdrive)
tesla1.drive()
tesla1.selfdriving()
```

When you execute tesla1.drive(), Python goes through this exact thought process:

It looks inside the Tesla class for a method named drive().

It sees that Tesla doesn't have one.

Because Tesla inherits from Car, Python then checks the Car parent class.

It finds def drive(self): in Car and runs it for you automatically.

You only need to write a method in Tesla if you want to override (change) how a parent action behaves. Otherwise, it inherits it perfectly for free.

```python
class Car:
    def __init__(self,windows,doors,engine_type):
        self.windows=windows
        self.doors=doors
        self.engine_type=engine_type
        
    def drive(self):
        print(f"The person will be driving {self.engine_type} car ")
        
class Tesla(Car):
    def __init__(self,is_selfdrive):
        super().__init__(windows=2,doors=4,engine_type="Electric")
        self.is_selfdrive=is_selfdrive
    def selfdriving(self):
        print(f"Tesla supports selfdriving {self.is_selfdrive}")
```

```python
tesla1=Tesla(True)
print(tesla1.windows)
print(tesla1.doors)
print(tesla1.is_selfdrive)
tesla1.drive()
tesla1.selfdriving()
```

Multiple Inheritance 
When a class inherits from more than one base class 

```python
#Creating BASE CLASS 1

class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("subclass implements this methods if not overwritten")

            
    
```

```python
# creating a base class 2 
class Owner:
    def __init__(self,owner):
        self.owner=owner
        
```

```python
#creating a derived class 
class Dog(Animal,Owner):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Owner.__init__(self,owner)
    def speak(self):
        return f"{self.name} says wooof"
    
```

```python
#creating object
dog1=Dog('Germanshepherd',"Abdullah")
print(dog1.name)
print(dog1.owner)
dog1.speak()
```

```python
# Base Class 1
class Animal:
    def __init__(self, name):
        self.name = name

# Base Class 2
class Owner:
    def __init__(self, owner):
        self.owner = owner

# Derived Class
class Dog(Animal, Owner):
    def __init__(self, name, owner):
        super(Dog, self).__init__(name) 
        super(Animal, self).__init__(owner) 

    def speak(self):
        return f"{self.name} says wooof"

# --- Testing the code ---
my_dog = Dog("Buddy", "Alex")
print(my_dog.name)   
print(my_dog.owner)  
```
