Polymorphism allow onjects of different classes to be treated as object of a common superclass


It is basically a way to perform a single action in different forms 

Achieved through two concepts Method overriding and Interfaces 

##Method Overriding 
Allows a child class to provide specific implementation of a method that is present in parent class

```python
class Animal:
    def speak(self):
        return "sound of an animal"
    
class Dog(Animal):
    def speak(self):
        return "dog sounds wooof"
        
class Cat(Animal):
    def speak(self):
        return "meow"
    
#function for demonstration 
def animal_speak(animal):
    print(animal.speak())
    
    
#creating objects 
dog1=Dog()
cat1=Cat()
print(dog1.speak())
print(cat1.speak())
animal_speak(dog1)
animal_speak(cat1)
    
```

```python
class Shape:
    def area(self):
        return "The area of figure"
    
class Rectangle(Shape):
    def __init__(self,length,width):
        self.length=length
        self.width=width
        
    def area(self):
        return self.length * self.width
    
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
        
    def area(self):
        return 3.14*self.radius*self.radius
    
#function for demonstration
def print_area(shape):
    print(f"area is {shape.area()}")


rect1=Rectangle(3,4)
circ1=Circle(3)

print_area(rect1)    
print_area(circ1)   
```

POLYMORPHISM THROUGH ABSTRACT BASE CLASS 

ABSTRACT BASE CLASS IS USED TO DEFINE A COMMON METHOD FOR GROUP OF RELATED OBJECTS. 

IT ENFORCES THAT DERIVED CLASS IMPLEMENTS PARTICULAR METHODS ,PROMOTING CONSISTENCY ACROSS DIFFERENT IMPLEMENTATIONS

It acts as a rigid blueprint or a contract. It forces all child classes to implement the exact same method names, ensuring they all speak the same "language" (interface).

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass  

class Car(Vehicle):
    def start_engine(self):
        return "Car engine purrs: Vroom!"

class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine roars: Braap!"


def ignition_sequence(vehicle_object):
    print(vehicle_object.start_engine())
    
my_car = Car()
my_bike = Motorcycle()

ignition_sequence(my_car)   
ignition_sequence(my_bike)  
```
