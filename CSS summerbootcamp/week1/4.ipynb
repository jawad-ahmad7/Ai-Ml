This section is all about filter function 
Filter function filters out items from a list (or an other iterable) based on condition

```python
def even(num):
    if num%2==0:
        print(f"{num} is an even entity")
        
        
```

```python
number=[1,2,3,4,5,6,7,8,9]
even=list(filter(even,number))
```

```python
# filer with a lamda function
number=[1,2,3,4,5,6,7,8,9]
even=list(filter(lambda x:x%2==0,number))
print(even)
```

```python
number=[1,2,3,4,5,6,7,8,9]
greater_than_five=list(filter(lambda x:x>5 ,number))
print(greater_than_five)
```

```python
number=[1,2,3,4,5,6,7,8,9]
greater_than_five_even=list(filter(lambda x:x>5 and x%2==0,number))
print(greater_than_five_even)
```

```python
#working withdictionaries using filter
people=[
    {'name':"abdullah",'roll_no':32},
    {'name':'ali','roll_no':1},
    {'name':'hamza','roll_no':9}
]
def roll_greater_than_five(person):
    return person["roll_no"]>5

age_greater=list(filter(roll_greater_than_five,people))
print(age_greater)
```
