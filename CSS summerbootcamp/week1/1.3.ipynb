This section is all conditional statements in python 

```python
#if statements
age=20
if age>=20:
    print("you are an adult")
```

```python
#else statement else will be executed if if statement is not true
age=17
if age>18:
    print("You can get a driving lincense ")
else:
    print("you need to wait")
```

```python
#elif statements
age=17
if age>=18:
    print("You are an adult get a driving lincense")
elif age==17:
    print('just wait another year to get a lincense')
else:
    print("you need to wait longer ")
```

```python
age = 25

if age < 10:
    print("Child ticket")        

elif age > 18:
    print("Adult ticket")       

elif age > 21:
    print("Senior discount")    
elif age == 25:
    print("Birthday special")   
```

```python
#nested if 
num=int(input("enter a number"))
if num>0:
    print(num,"is positive")
    if num%2==0:
        print(num,"is even")
    else:
        print(num,"is odd")
else:
    print(num,"is zero or negative")    
```

```python
#designing simple calculator 
num1=float(input("Enter the number"))
num2=float(input("Enter the number"))
op=str(input("Choose operation you wanna do +,*,/,%"))

if op == '+':
    result=num1 + num2
elif op== '-':
    result=num1 - num2
elif op== '*':
    result=num1 * num2
elif op== '/':
    result=num1 / num2
elif op== '%':
    result=num1 % num2
else:
    result='Invalid op'
    
print(result)
```
