```python
#packing in tuple
y=1,3,5,'abdullah'
packed_tuple=(y)
print(packed_tuple)
```

```python
#unpacking a tuple 
a,b,c,d=packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
#unpacking with *
numbers=(1,2,3,4,5,6,7,8)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
numbers=(1,2,3,4,5,6,7,8)
first*=numbers
print(first)
```
