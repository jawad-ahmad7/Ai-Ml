```python
#iteratingg over list
list=[]
for i in range(10):
    list.append(i)
    
print(list)    
```

```python
list=[]
for i in range(10):
    if i%2==0:
        a=list.append(i)

print(list)
```

```python
#list comprehension tchnique for same operation

a=[num for num in range(10) if num%2==0]
print(a)
```

BASIC SYNTAX = [EXPRESSION FOR ITEM IN ITERABLE]

Nested comprehension=[Expression for item in iterable 1 for item 2 in iterable 2 ]

```python
a=[num for num in range(10)]
print(a)
```

```python
b=[i**2 for i in range(10)]
print(b)
```
