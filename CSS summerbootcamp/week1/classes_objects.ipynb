The above method of adding attributes is not preferable because we loose blue print benefit in this case 
Imagine we are running a car factory, but your blueprint is just a blank piece of paper (pass).

For the first car, a worker decides to add 4 windows: audi.windows = 4.

For the second car, a different worker forgets and names it differently: bmw.win = 4.

For the third car, a worker forgets to add windows entirely.

Suddenly, our code becomes chaotic. If another part of your program tries to check bmw.windows, Python will crash

2. It Defeats "Encapsulation"
Remember how one of the main goals of OOP is to bundle data and behavior together?

When you add attributes on the fly outside the class, you are scattering your data definitions all over your program. If your project grows to 5,000 lines of code, you will have to hunt through different files just to figure out what data a Car object is supposed to hold.

```python
#so the preferrable way of adding attributes is using __init__ called constructor 
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
#creating objects 
dog1=Dog('buddy',13)
print(dog1.name)
print(dog1.age)
        
```

```python
#adding methods in above class 
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def bark(self):
        print(f'{self.name} barks like wooo')
        
        
#creating objects 
dog1=Dog('buddy',13)
print(dog1.name)
print(dog1.age)
dog1.bark()
```

```python
#modeling a  bank account 
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f"you have added {amount} in bank account now balance={self.balance}")
    def withdrawl(self,amount):
        if self.balance>amount:
            self.balance-=amount
            print(f'you had withdrawled {amount} now new balance is {self.balance}')
        else:
            print('Insufficent balance')
    def getbalance(self):
        print(f"your current account balance is {self.balance}")        
```

```python
#creating objects
acc1=BankAccount('Abdullah',5000)
print(acc1.owner)
print(acc1.balance)
```

```python
acc1.deposit(1000)
```

```python
acc1.withdrawl(1000)
```

```python
acc1.getbalance()
```
