This section is all about dictionaries 
Dictionaries are unordered,key value pairs,mutable,NOTE keys must be unique and are immutable 

```python
#creating a dictionary 
empty_dictionary=dict()
print(type(empty_dictionary))
```

```python
names={'name':'abdullah','roll_no':'32','address':'kohat'}
names
```

```python
#keys must be unique 
names={'name':'abdullah','roll_no':'32','address':'kohat','address':'peshawar'}
names 
```

```python
#accessing elements in dictionary
names={'name':'abdullah','roll_no':'32','address':'kohat'}
print(names['name'])
```

```python
print(names.get('roll_no'))
```

```python
#modifying dict elements 
names={'name':'abdullah','roll_no':'32','address':'kohat'}
names['name']='ali'
print(names)
```

```python
names['batch']='CS 2026'
print(names)
```

```python
#methods
names.keys()
```

```python
names.values()
```

```python
names.items()
```

```python
#shallow copy demo 
names_copy=names
print(names_copy)
print(names )

```

```python
names_copy['name']='abdullah'
print(names_copy)
print(names )
```

```python
#to remove this problem
names_copy=names.copy()
names_copy['name']='ali'
print(names_copy)
print(names )
```

```python
#iterating over dictionares 
names={'name': 'abdullah', 'roll_no': '32', 'address': 'kohat', 'batch': 'CS 2026'}
for i in names.keys():
    print(i)
```

```python
for i in names.values():
    print(i)
```

```python
for i,j in names.items():
    print(f"{i}:{j}")
```

```python
# creating nested dictionaries 
students={
    'student1':{'name':'abdullah','rollno':32},
    'student2':{'name':'ali','rollno':1}
}
print(students)
```

```python
#accssing elements in nested dict
print(students["student1"]['name'])
```

```python
print(students['student1'])
```

```python
#iterating over nested dict
for i,j in students.items():
    print(f"{i}:{j}")
    for key,value in j.items():
        print(f"{key}:{value}")
```

```python
#dict comprehension
sq={x:x**2 for x in range(10)}
print(sq)
```

```python
#conditional dict comprehension
sq={x:x**2 for x in range(10) if x%2==0}
print(sq)
```
