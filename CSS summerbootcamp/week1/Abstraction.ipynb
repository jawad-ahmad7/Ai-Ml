Abstraction is the cncept of hiding complex implementation details and showing only necessary features of an object 

```python
from abc import ABC, abstractmethod 

class Vehicle(ABC): 
    def drive(self): 
        print('vehicle is used to drive') 

    @abstractmethod 
    def start_engine(self): 
        pass 

class Car(Vehicle): 
    def start_engine(self): 
        print("car engine started") 

def operate_vehicle(vehicle): 
    vehicle.start_engine() 
    vehicle.drive() 

car = Car() 
operate_vehicle(car)
```

```python
from abc import ABC, abstractmethod

# 1. THE BLUEPRINT (Abstraction)
# This shows ONLY the necessary feature: "pay"
class PaymentProcessor(ABC):
    
    @abstractmethod
    def pay(self, amount):
        pass  # Starts empty! Just a placeholder button.

# 2. HIDDEN COMPLEX IMPLEMENTATION (Credit Card)
class CreditCardPayment(PaymentProcessor):
    def pay(self, amount):
        # 50 lines of complex hidden logic would go here:
        # Connect to Visa/Mastercard network -> Check CVV -> Verify credit limit
        print(f"Processing credit card payment of ${amount} via bank servers...")

# 3. HIDDEN COMPLEX IMPLEMENTATION (PayPal)
class PayPalPayment(PaymentProcessor):
    def pay(self, amount):
        # Different complex hidden logic:
        # Open API connection -> Log into PayPal user wallet -> Transfer tokens
        print(f"Processing PayPal payment of ${amount} via digital wallet...")


# 4. THE USER (Doesn't care about the complex machinery)
def checkout_website(payment_method, total_price):
    # The website only needs to know the "pay" button exists
    payment_method.pay(total_price) 


# --- Running the code ---
card_payment = CreditCardPayment()
paypal_payment = PayPalPayment()

# The website handles both payments exactly the same way!
checkout_website(card_payment, 100)
checkout_website(paypal_payment, 50)
```
