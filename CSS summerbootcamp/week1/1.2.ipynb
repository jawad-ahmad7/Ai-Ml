This section is all about arthimetic operators,logical operators,comparison operators 

```python
## arthimetic operators 
a=10;b=5
addition=a+b
subtraction=a-b
division=a/b
multiplication=a*b
floor=a//b
power=a**b
modulus=a%b

print(addition)
print(subtraction)
print(division)
print(multiplication)
print(floor)
print(power)
print(modulus)
```

```python
#logical operators 
a=True
b=False
print(a and b)
print(a or b)
print(not b) 
```

```python
# comparison operators
a='Abdullah'
b='Abdullah'
a==b
```

```python
a=10
b=9
print(a==b)
print(a>b)
print(a>=b)
print(a<=b)
print(a<b)
```
