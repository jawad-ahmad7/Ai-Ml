This section is all about lambda function 
Lambda function is an anynomous function 
Takes any no of arguments but only one expression 
used for short operations 
or as an arguments to high order functions

#syntax
lamda parameter: expressions

```python
# for example 
def add(a,b):
    return a+b

add(1,2)
```

```python
add=lambda a,b:a+b
print(add(1,2))
```

```python
def even(num):
    if num%2==0:
        print(f"{num} is  even ")
        
even(24)
```

```python
even=lambda num:num%2==0
print(even(24))
```

```python
#working with more parameters in lamda 
add1=lambda x,y,z:x+y+z
print(add1(1,2,3))
```
