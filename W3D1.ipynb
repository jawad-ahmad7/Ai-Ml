### circle class 

```python
class Circle:
    def __init__(self,radius):
        self.radius=radius
    def cal_area(self):
        print(f"area of circle is : {3.14*self.radius**2}")
    def cal_circuference(self):
        print(f"circuference of circle is : {2*3.14*self.radius  }")
       
```

```python
circle=Circle(5)
```

```python
circle.cal_area()
```

```python
circle.cal_circuference()
```

### student class

```python
class Student:
    def __init__(self,id,name,grade_level):
        self.id=id
        self.name=name
        self.grade_level=grade_level
        self.courses=[]

    def enroll_in_course(self,course):
        if course in self.courses:
            print(f"yoy are enrollrd in {course}")
        else:
            self.courses.append(course)
            print("successful enrollment")
          
```

```python
student1=Student(1,"jawad",7)
```

```python
student1.courses=["python","java","c++"]
print(student1.courses)
```

```python
student1.enroll_in_course("python")
```
